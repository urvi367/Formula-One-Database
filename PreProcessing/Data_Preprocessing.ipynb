```python
import pandas as pd

circuits_df = pd.read_csv("circuits.csv")
circuits_df.head()
```

```python
selected_columns_df = circuits_df[['circuitId', 'name', 'location', 'country']]
selected_columns_df.head()
```

```python
def df_to_sql_insert(df, table_name):

    sql_statements = []
    for index, row in df.iterrows():
        formatted_values = []
        for item in row:
            if isinstance(item, str):
                formatted_values.append("'" + item.replace("'", "''") + "'")
            else:
                formatted_values.append(str(item))

        values_str = ", ".join(formatted_values)
        sql_statement = f"({values_str})"
        sql_statements.append(sql_statement)

    full_sql = f"INSERT INTO {table_name} VALUES\n    " + ",\n    ".join(sql_statements) + ";"
    return full_sql
```

```python
sql_insert_statements = df_to_sql_insert(selected_columns_df, "Circuits")
sql_insert_statements
```

```python
sql_file_path = 'circuits.sql'
with open(sql_file_path, 'w') as file:
    file.write(sql_insert_statements)

sql_file_path
```
